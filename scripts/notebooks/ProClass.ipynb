### Task 3 (Arithmetic Operator):

- Take an input of +,-,/,*,and % from user and compute arithmetic operations using match case.
  - Match Case in Python is a pattern-matching feature that works as switch case in C/C++. Will Study more in further incoming course.
  - Use while() loop if possible. Here, In python do_while doesnt exist liek C/C++.
  

```python
#Take an input of +,-,/,*,and % from user and compute arithmetic operations.
num1 = int(input("Input First Number"))
num2 = int(input("Input Second Number"))
while(True):
    sym = input("Input Operations : + , - , / , * , % ,e for exit")
    match sym:
    
        case '+':
            add=num1+num2
            print(f"Addition is {add}")
        case '-':
            diff=num1-num2
            print(f"Difference is {diff}")
        case '/':
            div=num1/num2
            print(f"division is {div}")
        case '*':
            prod=num1*num2
            print(f"Product is {prod}")
        case '%':
            mod=num1%num2
            print(f"Remainder is {mod}")
        case 'e':
            print("Program Exited")
            break
        case _:
            print("Invalid Operator")


        

    
    


```

```python
###Task 4
- Create an online voting system.
- User Inputs:
    Age,Name,Party choice
- Use relational operator to check voting eligibility
- Party option
    - Party A
    - Party B
    - Party C
    - Party D

- Display message: your vote has been casted successfully
- Invalid Message
```

## Data Structures in Python

- It specify how the data are stored in program.
- There are **Four** data structures:
    - List :
        - It is similar to array in other programming language which is the **ordered collection** of data
        - We access elements using index.
        - It is dynamic list
        - It can handle multiple datatypes
        - Syntax **listname = []**
        - It is **mutable** i.e we can modify it in nature.
        - It allows duplication (we can element in multiple index)
    - Tuple
        - Tuple are similar to list but are **immutable** (we cannot change)
    	- Tuple cannot be modified once it is declared
    	- Syntax: **tuple_name = (element1, element2, elemennt3)**
    - Dictionary
    - Set 


```python
## Empty List

## List with same datatype

list1=[]
list1['a','b','c','d','e']
print(list1)

## List with different datatype

list2 = [1,1.5,"Texas","College"]
print(list2)

## Duplication in list

list3 = [1,1,2,3,4,3,3]
print(list3)

## Access Elements in list

print(list2[3])
print(list2[2:]) #It Display elements from 2nd Position ignoring 0 & 1 index.

## Python Support negative indexing

print(list2[-1])
print(list2[-1:]) 
print("It displays elements from last position")
```
