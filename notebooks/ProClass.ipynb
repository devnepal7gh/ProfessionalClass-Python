### Task 2:

- Take input of two numbers from user and swap them
    - Using temp variable

```python
##Using Temporary Variable
number1=int(input("Enter First Number"))
number2=int(input("Enter Second Numeber"))
print("Before Swap:")
print(f"Number1 = {number1}")
print(f"Number2 = {number2}")
print("\n")
print("Using Temp Variable")
tempNum=(number1)
number1=number2
number2=tempNum
print("After Swap:")
print(f"Number1={number1}")
print(f"Number2={number2}")
print("\n")
##Without Using Temporary Variable
print("Without Using Temp Variable")
print("Before Swap:")
print(f"Number1 = {number1}")
print(f"Number2 = {number2}")
num3=int(number1+number2)
number1=num3-number1
number2=num3-number2
print("After Swap:")
print(f"Number1={number1}")
print(f"Number2={number2}")
print("\n")
##Python Swap

print("Python Swap")
print("Before Swap:")
print(f"Number1 = {number1}")
print(f"Number2 = {number2}")
(number1,number2)=(number2,number1)
print("After Swap:")
print(f"Number1={number1}")
print(f"Number2={number2}")
    
            
```
