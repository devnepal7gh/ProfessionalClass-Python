### Task 3 (Arithmetic Operator):

- Take an input of +,-,/,*,and % from user and compute arithmetic operations using match case.
  - Match Case in Python is a pattern-matching feature that works as switch case in C/C++. Will Study more in further incoming course.
  - Use while() loop if possible. Here, In python do_while doesnt exist liek C/C++.
  

````python
#Take an input of +,-,/,*,and % from user and compute arithmetic operations.
num1 = int(input("Input First Number"))
num2 = int(input("Input Second Number"))
while(True):
    sym = input("Input Operations : + , - , / , * , % ,e for exit")
    match sym:
    
        case '+':
            add=num1+num2
            print(f"Addition is {add}")
        case '-':
            diff=num1-num2
            print(f"Difference is {diff}")
        case '/':
            div=num1/num2
            print(f"division is {div}")
        case '*':
            prod=num1*num2
            print(f"Product is {prod}")
        case '%':
            mod=num1%num2
            print(f"Remainder is {mod}")
        case 'e':
            print("Program Exited")
            break
        case _:
            print("Invalid Operator")


        

    
    


````
