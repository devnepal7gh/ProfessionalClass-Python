```python
from abc import ABC, abstractmethod
import random


class BankAccount(ABC):
    def __init__(self, name, balance, pin):
        self.name = name
        self.__balance = balance
        self.__pin = pin
        self.accountNumber = self._generateAccountNo()
        print(f"Welcome {self.name}! Your Account has been created")

    def _generateAccountNo(self):
        return str(random.randint(100000, 999999))

    def verifyPin(self, user_pin):
        return self.__pin == user_pin

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid amount")
            return
        self.__balance += amount
        print(f"${amount} deposited")
        self.checkbalance()

    def checkbalance(self):
        print(f"Your total balance is {self.__balance}")

    def _getBalance(self):
        return self.__balance

    def _deductBalance(self, amount):
        self.__balance -= amount

    @abstractmethod
    def withdraw(self, amount):
        pass


class SavingAccount(BankAccount):
    def __init__(self, name, balance, pin):
        super().__init__(name, balance, pin)

    def withdraw(self, amount):
        if amount <= 0:
            print("Amount must be greater than 0")
            return

        if self._getBalance() >= amount:
            self._deductBalance(amount)
            print(f"{amount} withdrawn")
            self.checkbalance()
        else:
            print("Insufficient balance")


class CheckingAccount(BankAccount):
    def __init__(self, name, balance, pin, overdraftLimit):
        super().__init__(name, balance, pin)
        self.overdraftLimit = overdraftLimit

    def withdraw(self, amount):
        if amount <= 0:
            print("Amount must be greater than 0")
            return

        available = self._getBalance() + self.overdraftLimit

        if available >= amount:
            self._deductBalance(amount)
            print(f"{amount} withdrawn")
            self.checkbalance()
        else:
            print("Exceeded overdraft limit")


def main():
    accountDb = {}
    while(True):
        print("========= Welcome to Python Bank Pvt. ltd. =========")
        print("1. Open New Acoount")
        print("2. Login to Existing Account")
        print("3. Exit")

        choice = input("Enter your choice:")
        if (choice == "1"):
            name = input("ENter Your name:")
            pin = input("Enter Your 4 Digit Pin")

            try:
                balance = float(input("Enter Your initial Balance"))
            except ValueError as e:
                print("Value Error",e)
                continue

            print("Choose Account Type:")
            print("S - Saving Account")
            print("C - Checking Account")

            accType = input("Enter Your account Type:")
            if accType == 'S':
                new_acc = SavingAccount(name,balance,pin)
                print(new_acc.accountNumber)
            elif accType == 'C':
                new_acc = CheckingAccount(name,balance,pin,1000)
                print(new_acc.accountNumber)
            else:
                print ("Invalid Choice")
                continue

            accountDb[new_acc.accountNumber] = new_acc

        elif(choice == '2'):
            accNo = input("Enter the account Number")
            if accNo in accountDb:
                currentAcc = accountDb[accNo]
                user_pin = input("Enter Your Pin")

                if(currentAcc.verifyPin(user_pin)):
                    print("Welcome back to Banking System")

                    while True:
                        print("1. For Deposit")
                        print("2. For Withdraw")
                        print("3. For Display")
                        print("4. For Logout")

                        user_action = input("Enter Your Option")
                        if user_action == '1':
                            amt = float(input("Enter Your Amount"))
                            currentAcc.deposit(amt)
                        elif user_action == '2':
                            amt = float(input("enterr Your Amount"))
                            currentAcc.withdraw(amt)
                        elif user_action == '3':
                            print(f"Welcome {currentAcc.name}")
                            currentAcc.checkbalance
                        elif user_action == '4':
                            print("Logging out of the system")
                            break
                        else:
                            print("Invalid Input")
                            continue
                    
            else:
                print("Input Valid Account Number")
                    
        elif (choice == '3'):
            print("ThankYou for banking with python")
            break

        else:
            print("Invalid Input. Try Again")

main()
    
            
            
            
```
